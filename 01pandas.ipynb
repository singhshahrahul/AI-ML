## Understanding Mean, Median, and Mode

These are the three main measures of central tendency used in statistics:

1.  **Mean (Average)**: The sum of all values divided by the total number of values.
    *   *Example*: For [1, 2, 3], the mean is (1+2+3)/3 = 2.
2.  **Median (Middle)**: The middle value when the data is ordered from smallest to largest. If there is an even number of values, it is the average of the two middle numbers.
    *   *Example*: For [1, 2, 100], the median is 2 (the mean would be 34.3, showing how the median is less affected by outliers).
3.  **Mode (Most Frequent)**: The value that appears most often in a dataset.
    *   *Example*: In [Toyota, Honda, Toyota], the mode is Toyota.

```python
# Demonstration using the Car Sales Odometer data
odometer_data = car_sales['Odometer (KM)']

print(f"Odometer Mean: {odometer_data.mean()}")
print(f"Odometer Median: {odometer_data.median()}")
print(f"Odometer Mode: {car_sales['Make'].mode()[0]} (Most frequent car make)")
```
