```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

#Workflow

```python
#1. prepare data
x = [1,2,3,4]
y = [2,4,6,8]

#2. setup plot
fig, ax = plt.subplots(figsize=(5,5))

#3.plot data
ax.plot(x, y)

#4. customize plot
ax.set(title = "Simple plot", xlabel = "X-axis", ylabel = "Y-axis")

#5. save & show
fig.savefig("sample-plot")
```

#Two option for subplot

```python
#option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.hist(np.random.randn(100000))
ax4.bar(car_prices.keys(), car_prices.values());
```

```python
#option 2
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10,5))
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].hist(np.random.randn(100000))
ax[1,1].bar(car_prices.keys(), car_prices.values());
```

# 1st method from scratch

```python
fig, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=car_sales["Price"], y=car_sales["total sales"])

#customize the plot
ax.set(title="cars", xlabel="price", ylabel="sales")

#add legend
ax.legend(*scatter.legend_elements(), title="Doors")

#add horizontal line
ax.axhline(car_sales["Price"].mean(), linestyle="--");
```

```python
hd = pd.read_csv("/content/heart-disease.csv")
over_50 = hd[hd["age"]>50]
over_50.head()
```

```python
plt.style.use("seaborn-v0_8-whitegrid")
#Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(ncols = 1,
                               nrows = 2,
                               figsize =(10,10), sharex=True)

#add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"],
                      c=over_50["target"], cmap="winter") #change the color of dots

#customize ax0
ax0.set(title= "Heart Disease and Cholesterol level", ylabel = "chol")

#add legend
ax0.legend(*scatter.legend_elements(), title="target")

#add a mean line
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

#add data to ax1
scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"],
                      c=over_50["target"], cmap="summer")

#customize ax1
ax1.set(title= "Heart Disease and Max heart rate", xlabel="age", ylabel = "max heart rate")

#change ax1 x-axis limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

#add legend
ax1.legend(*scatter.legend_elements(), title="target");

#add a title for figure
fig.suptitle("Heart Dicease Analysis", fontsize=16, fontweight="bold");
```
