In Python, strings are **immutable**. This means that once a string is created, its content cannot be changed. Any operation that appears to modify a string, like replacement or concatenation, actually creates a new string.

````python
my_string = "Python"
print(f"Original string: {my_string}")

# Trying to change a character will result in an error
# my_string[0] = 'J' # This would raise a TypeError

# Operations like replace create a new string
new_string = my_string.replace('P', 'J')
print(f"New string after replace: {new_string}")
print(f"Original string is still: {my_string}") # Original string remains unchanged
````

### 10. More Essential String Methods

Here are some other commonly used string methods that are very useful, especially in data preprocessing for Machine Learning:

````python
# .strip() - Removes leading and trailing whitespace
text = "  Hello, World!  "
stripped_text = text.strip()
print(f"strip(): '{stripped_text}'")

# .lower() - Converts all characters to lowercase
message = "HELLO Python"
lower_message = message.lower()
print(f"lower(): '{lower_message}'")

# .upper() - Converts all characters to uppercase
statement = "this is a test"
upper_statement = statement.upper()
print(f"upper(): '{upper_statement}'")

# .split() - Splits the string into a list of substrings based on a delimiter
sentence = "Python is a versatile language"
words = sentence.split()
print(f"split() by default (whitespace): {words}")

data = "apple,banana,cherry"
fruits = data.split(',')
print(f"split(',') with comma delimiter: {fruits}")

# .join() - Joins elements of an iterable (like a list) with the string as a separator
word_list = ['Join', 'these', 'words']
joined_string = ' '.join(word_list)
print(f"join() with space separator: '{joined_string}'")

dash_separated = '-'.join(['file', 'name', 'txt'])
print(f"join() with hyphen separator: '{dash_separated}'")
````

### 4.5 Lists vs. Tuples: Key Differences

While both lists and tuples are used to store collections of items, they have fundamental differences that dictate their use cases:

*   **Mutability:**
    *   **Lists:** Are **mutable**, meaning their elements can be changed (added, removed, modified) after creation. This makes them suitable for collections of items that may need to be updated.
    *   **Tuples:** Are **immutable**, meaning their elements cannot be changed after creation. This makes them ideal for fixed collections of items or as dictionary keys (since keys must be immutable).

*   **Syntax:**
    *   **Lists:** Defined using square brackets `[]`.
    *   **Tuples:** Defined using parentheses `()`.

*   **Performance:**
    *   Tuples are generally slightly faster than lists for iteration and access because their size is fixed.

*   **Use Cases:**
    *   **Lists:** Use when you expect the collection to change (e.g., a list of users, a shopping cart).
    *   **Tuples:** Use when you need a fixed collection of items (e.g., coordinates, RGB color values, database records) or as arguments to functions where the data should not be altered.

## **DICTIONARIES**

Dictionaries in Python are unordered collections of data values, used to store data values like a map, which, unlike other Data Types that hold only a single value as an element, holds `key:value` pairs. Each `key:value` pair in a Dictionary is separated by a comma, and the entire dictionary is enclosed in curly braces `{}`.

### 5.1 Creating and Accessing Dictionaries

Dictionaries are mutable, meaning they can be changed after they are created.

````python
# Creating an empty dictionary
my_dict = {}
print(f"Empty dictionary: {my_dict}")

# Creating a dictionary with key-value pairs
student = {
    'name': 'Alice',
    'age': 20,
    'major': 'Computer Science',
    'courses': ['Math', 'Physics', 'Programming']
}
print(f"Student dictionary: {student}")

# Accessing values using keys
print(f"Student's name: {student['name']}")
print(f"Student's age: {student.get('age')}") # Using .get() for safer access

# Trying to access a non-existent key with .get() (returns None by default)
print(f"Student's GPA (using get): {student.get('gpa')}")
# Accessing a non-existent key directly will raise a KeyError
# print(student['gpa'])
````

### 5.2 Modifying Dictionaries

You can add new key-value pairs or update existing ones.

````python
student = {
    'name': 'Alice',
    'age': 20,
    'major': 'Computer Science'
}

# Adding a new key-value pair
student['gpa'] = 3.8
print(f"After adding GPA: {student}")

# Updating an existing value
student['age'] = 21
print(f"After updating age: {student}")

# Using update() to add multiple items or update existing ones
student.update({
    'city': 'New York',
    'age': 22 # Updates age again
})
print(f"After update() method: {student}")
````

### 5.3 Removing Elements from Dictionaries

Elements can be removed using `del`, `pop()`, or `clear()`.

````python
student = {
    'name': 'Alice',
    'age': 22,
    'major': 'Computer Science',
    'gpa': 3.8,
    'city': 'New York'
}

# Removing a key-value pair using del
del student['city']
print(f"After deleting 'city': {student}")

# Removing a key-value pair using pop() (returns the value)
removed_gpa = student.pop('gpa')
print(f"Removed GPA: {removed_gpa}")
print(f"After pop('gpa'): {student}")

# Using popitem() to remove and return an arbitrary (last inserted in Python 3.7+) key-value pair
last_item = student.popitem()
print(f"Removed item with popitem(): {last_item}")
print(f"After popitem(): {student}")

# Clearing all items from the dictionary
student.clear()
print(f"After clear(): {student}")
````

### 5.4 Dictionary Methods

Important methods to retrieve keys, values, and items.

````python
student = {
    'name': 'Bob',
    'age': 25,
    'major': 'Physics',
    'gpa': 3.5
}

# Get all keys
keys = student.keys()
print(f"Keys: {keys}, Type: {type(keys)}")

# Get all values
values = student.values()
print(f"Values: {values}, Type: {type(values)}")

# Get all items (key-value pairs)
items = student.items()
print(f"Items: {items}, Type: {type(items)}")

# Iterating through a dictionary
print("\nIterating through items:")
for key, value in student.items():
    print(f"{key}: {value}")
````

### 5.5 Dictionary Comprehensions

Similar to list comprehensions, dictionary comprehensions provide a concise way to create dictionaries.

````python
# Create a dictionary of squares
squares = {num: num**2 for num in range(5)}
print(f"Squares dictionary: {squares}")

# Another simple example: Create a dictionary where keys are fruits and values are their length

fruits = ['apple', 'banana', 'cherry']
fruit_lengths = {fruit: len(fruit) for fruit in fruits}
print(f"Fruit lengths dictionary: {fruit_lengths}")

# Create a dictionary from two lists
keys = ['a', 'b', 'c']
values = [1, 2, 3]
my_dict_from_lists = {k: v for k, v in zip(keys, values)}
print(f"Dictionary from lists: {my_dict_from_lists}")
````

## **SETS**

Sets are unordered collections of unique elements. They are mutable, but the elements within a set must be immutable (like numbers, strings, tuples).

### 5.6 Creating and Modifying Sets

Sets are useful for mathematical set operations like union, intersection, and difference, and for efficiently checking for membership.

````python
# Creating an empty set
empty_set = set() # Note: {} creates an empty dictionary, not an empty set
print(f"Empty set: {empty_set}")

# Creating a set with elements (duplicates are automatically removed)
my_set = {1, 2, 3, 2, 4, 1}
print(f"Set with unique elements: {my_set}")

# Creating a set from a list
list_to_set = set([5, 6, 7, 6])
print(f"Set from a list: {list_to_set}")

# Adding elements to a set
my_set.add(5)
print(f"After adding 5: {my_set}")
my_set.add(1) # Adding an existing element does nothing
print(f"After adding 1 again: {my_set}")

# Adding multiple elements using update()
my_set.update([6, 7, 8])
print(f"After updating with a list: {my_set}")
````

### 5.7 Removing Elements from Sets

Elements can be removed using `remove()`, `discard()`, `pop()`, or `clear()`.

````python
my_set = {1, 2, 3, 4, 5, 6}

# Removing an element using remove()
my_set.remove(3)
print(f"After removing 3: {my_set}")
# my_set.remove(10) # This would raise a KeyError if element not found

# Removing an element using discard()
my_set.discard(2)
print(f"After discarding 2: {my_set}")
my_set.discard(10) # Discarding a non-existent element does nothing
print(f"After discarding 10 (non-existent): {my_set}")

# Removing an arbitrary element using pop()
popped_element = my_set.pop()
print(f"Popped element: {popped_element}")
print(f"After pop(): {my_set}")

# Clearing all elements from the set
my_set.clear()
print(f"After clear(): {my_set}")
````

### 5.8 Set Operations

Sets support mathematical operations like union, intersection, difference, and symmetric difference.

````python
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

print(f"Set A: {set_a}")
print(f"Set B: {set_b}")

# Union: All unique elements from both sets
union_set = set_a.union(set_b) # or set_a | set_b
print(f"Union (A | B): {union_set}")

# Intersection: Common elements in both sets
intersection_set = set_a.intersection(set_b) # or set_a & set_b
print(f"Intersection (A & B): {intersection_set}")

# Difference: Elements in A but not in B
difference_set_ab = set_a.difference(set_b) # or set_a - set_b
print(f"Difference (A - B): {difference_set_ab}")

# Difference: Elements in B but not in A
difference_set_ba = set_b.difference(set_a) # or set_b - set_a
print(f"Difference (B - A): {difference_set_ba}")

# Symmetric Difference: Elements in either A or B, but not both
symmetric_difference_set = set_a.symmetric_difference(set_b) # or set_a ^ set_b
print(f"Symmetric Difference (A ^ B): {symmetric_difference_set}")

# Subset and Superset
set_c = {1, 2}
print(f"Is {set_c} a subset of {set_a}? {set_c.issubset(set_a)}")
print(f"Is {set_a} a superset of {set_c}? {set_a.issuperset(set_c)}")
````

#CHAPTER 6 `CONDITIONAL EXPRESSION`

Conditional expressions, or `if` statements, are fundamental in programming for controlling the flow of execution based on certain conditions. They allow your program to make decisions and execute different blocks of code depending on whether a condition is `True` or `False`.

### 6.1 The `if` Statement

The `if` statement is the simplest form of a conditional. It executes a block of code only if a specified condition is `True`.

````python
score = 75

if score >= 70:
    print("Congratulations! You passed.")

print("End of program.")

score = 60
if score >= 70:
    print("Congratulations! You passed.")

print("End of program (second check).")
````

### 6.2 The `if-else` Statement

The `if-else` statement allows you to specify a block of code to be executed if the condition is `True`, and an alternative block to be executed if the condition is `False`.

````python
age = 18

if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not yet eligible to vote.")

price = 100
discount_applied = False

if price > 50:
    discount_amount = price * 0.10
    final_price = price - discount_amount
    discount_applied = True
    print(f"Discount applied! Final price: ${final_price:.2f}")
else:
    final_price = price
    print(f"No discount. Final price: ${final_price:.2f}")

print(f"Was discount applied? {discount_applied}")
````

When you have multiple conditions to check, you can use the `if-elif-else` chain. The `elif` (short for 'else if') keyword allows you to check multiple expressions for `True` and execute a block of code as soon as one of the conditions evaluates to `True`.

### 8.5 Introduction to Recursion
Recursion is a process where a function calls itself. It is often used to solve mathematical problems like factorials or Fibonacci sequences.

**Crucial Note:** A recursive function must have a **Base Case** to stop it from calling itself infinitely.

````python
def factorial_recursive(n):
    # Base case: factorial of 0 or 1 is 1
    if n == 0 or n == 1:
        return 1
    # Recursive case
    else:
        return n * factorial_recursive(n - 1)

num = 5
print(f"Factorial of {num} using recursion is: {factorial_recursive(num)}")
````

### 8.6 How Recursion Works (Step-by-Step)

When you call `factorial_recursive(5)`, the function doesn't return a result immediately. It keeps calling itself until it hits the **base case**. Here is the mental model of the process:

**1. The Expansion Phase (Going Down):**
- `factorial_recursive(5)` returns `5 * factorial_recursive(4)`
- `factorial_recursive(4)` returns `4 * factorial_recursive(3)`
- `factorial_recursive(3)` returns `3 * factorial_recursive(2)`
- `factorial_recursive(2)` returns `2 * factorial_recursive(1)`
- `factorial_recursive(1)` returns `1` **(Base Case reached!)**

**2. The Substitution Phase (Coming Up):**
- Now the values start returning back up the chain:
- `factorial_recursive(1) = 1`
- `factorial_recursive(2) = 2 * 1 = 2`
- `factorial_recursive(3) = 3 * 2 = 6`
- `factorial_recursive(4) = 4 * 6 = 24`
- `factorial_recursive(5) = 5 * 24 = 120`

# Chapter 9 `FILE I/O`

Files are used to store data permanently. In Python, we use the `open()` function to handle files.

### 9.1 Opening a File and Modes

When opening a file, we specify a **mode**:
- `'r'`: **Read mode** (default). Opens a file for reading; error if the file doesn't exist.
- `'w'`: **Write mode**. Opens a file for writing; creates it if it doesn't exist or **overwrites** it if it does.
- `'a'`: **Append mode**. Opens a file for appending data at the end without deleting existing content.
- `'rb'` / `'wb'`: Used for **Binary files** (like images or PDFs).

````python
# 1. Writing to a file
f = open("myfile.txt", "w")
f.write("Hello, this is Chapter 9.\n")
f.write("Learning File I/O is essential.")

# 2. Closing the file
f.close()

print("File created and closed successfully.")
````

### ⚠️ CRITICAL: The Behavior of Write Mode (`'w'`)

It is vital to understand that when you open a file in `'w'` mode, Python **immediately wipes out** everything that was previously in that file before you even write a single line.

- **'w' (Write):** Deletes existing content and starts fresh.
- **'a' (Append):** Keeps existing content and adds new data to the end of the file.

Always double-check your mode before running your code!

### 9.2 Why do we use `f.close()`?

It is crucial to close a file after you are done with it because:
1. **Resource Management:** Operating systems have a limit on how many files can be open at once.
2. **Data Integrity:** Closing a file ensures that all buffered data is physically written to the disk.
3. **Prevents Corruption:** Leaving files open can lead to data loss if the program crashes.

### 9.3 Reading a File

We can read the entire content using `.read()`, a single line using `.readline()`, or all lines into a list using `.readlines()`.

````python
f = open("myfile.txt", "r")
content = f.read()
print("--- File Content ---")
print(content)
f.close()
````

The `with` statement is the most optimized and safest way to handle files.

**Why use it?**

# Chapter 10 `OBJECT ORIENTED PROGRAMMING`

### 10.1 Introduction to OOP
Solving a problem by creating objects is one of the most popular approaches in programming. This is called Object-Oriented Programming.

**Dry Principle:** OOP helps us follow the 'Don't Repeat Yourself' principle. Instead of writing separate variables and functions for every user or item, we create a template (Class) and reuse it.

### 10.2 Class and Object
- **Class:** A blueprint for creating objects (e.g., a blank Application Form).
- **Object:** An instantiation of a class (e.g., a form filled with 'John's' details).

````python
class Employee:
    language = "Python" # Class Attribute
    salary = 1200000

emp1 = Employee()
emp1.name = "John"
print(emp1.name, emp1.language, emp1.salary)
````

### 10.3 The `self` Parameter
`self` refers to the instance of the class. It is automatically passed with any function call from an object. It allows the function to access attributes and methods of that specific object.

````python
class Employee:
    language = "Python"
    salary = 1200000

    def getDetails(self):
        print(f"The language is {self.language}. The salary is {self.salary}")

emp1 = Employee()
emp1.getDetails() # This is equivalent to Employee.getDetails(emp1)
````

### 10.4 The `__init__` Constructor
`__init__()` is a special method which is run as soon as an object is created. It is used to initialize the attributes of an object.

````python
class Employee:
    def __init__(self, name, salary, language):
        self.name = name
        self.salary = salary
        self.language = language
        print("I am creating an object")

rohan = Employee("Rohan", 1300000, "JavaScript")
print(rohan.name, rohan.salary, rohan.language)
````

### 10.5 Static Methods (`@staticmethod`)
Sometimes we need a function inside a class that doesn't use any data from the object (doesn't use `self`). For such cases, we use the `@staticmethod` decorator. It allows us to call the method without passing the instance automatically.

````python
class Calculator:
    def __init__(self, num):
        self.num = num

    @staticmethod
    def greet():
        print("Hello! Welcome to the Calculator.")

# We can call static methods without creating an object,
# or using an object without passing 'self'
Calculator.greet()

calc = Calculator(10)
calc.greet()
````

# Chapter 11 `INHERITANCE & ADVANCED OOP`

### 11.1 What is Inheritance?
Inheritance is a way of creating a new class from an existing one. The new class (Child/Derived class) inherits the attributes and methods of the existing class (Parent/Base class).

**Benefits:**
- Code Reusability.
- Logical Structure.

**Syntax:**
```python
class Parent:
    # Parent body

class Child(Parent):
    # Child body inherits Parent
```

````python
class Employee:
    company = "Google"
    def showDetails(self):
        print("This is an employee")

class Programmer(Employee):
    language = "Python"
    def getLanguage(self):
        print(f"The language is {self.language}")

p = Programmer()
print(p.company) # Inherited attribute
p.showDetails()  # Inherited method
p.getLanguage() # Own method
````

### 11.2 Types of Inheritance
1. **Single Inheritance:** Derived class inherits from a single base class.
2. **Multiple Inheritance:** Derived class inherits from more than one base class.
3. **Multilevel Inheritance:** A derived class becomes the base class for another class.

````python
class Father:
    def trait1(self):
        print("Focus from Father")

class Mother:
    def trait2(self):
        print("Creativity from Mother")

class Child(Father, Mother):
    def show(self):
        print("Child inherits from both")

c = Child()
c.trait1()
c.trait2()
c.show()
````

### 11.3 The `super()` Method
The `super()` method is used to call methods of the parent class inside a child class. It is commonly used within `__init__` to ensure the parent class is properly initialized.

````python
class Person:
    def __init__(self):
        print("Initializing Person")

class Employee(Person):
    def __init__(self):
        super().__init__() # Calls Parent constructor
        print("Initializing Employee")

e = Employee()
````

### 11.4 Class Methods (`@classmethod`)
A class method is bound to the class and not the object. It can modify a class state that applies across all instances. It takes `cls` as the first argument instead of `self`.

````python
class Employee:
    salary = 100

    @classmethod
    def changeSalary(cls, sal):
        cls.salary = sal

e = Employee()
print(f"Old Salary: {Employee.salary}")
e.changeSalary(200)
print(f"New Salary: {Employee.salary}")
````

### 12.1 Modern Python Features
Python is constantly evolving. Here are some of the most useful features added in recent versions that make your code shorter and more readable.

#### 1. The Walrus Operator (`:=`)
Introduced in Python 3.8, it allows you to **assign** a value to a variable **inside** an expression (like an `if` statement or a `while` loop).

````python
# Example: Check length and store it at the same time
if (n := len("Python")) > 3:
    print(f"The word has {n} characters.")
````

#### 2. Type Hints
Type hints help other developers (and your code editor) understand what kind of data a function expects and returns. It doesn't stop the code from running with wrong types, but it makes it much easier to debug.

````python
def calculate_total(price: float, tax: float) -> float:
    return price + tax

print(calculate_total(100.0, 18.5))
````

#### 3. Match Case (Structural Pattern Matching)
Added in Python 3.10, this is like a more powerful version of `if-elif-else` for matching specific patterns or values.

````python
def get_day_type(day):
    match day:
        case "Saturday" | "Sunday":
            return "Weekend"
        case "Monday" | "Tuesday" | "Wednesday" | "Thursday" | "Friday":
            return "Weekday"
        case _:
            return "Invalid day"

print(get_day_type("Sunday"))
````

### 12.2 Exception Handling
Exceptions are errors that happen during execution. Instead of letting your program crash, you can 'catch' these errors.

````python
try:
    num = int(input("Enter a number to divide 100: "))
    result = 100 / num
    print(f"Result: {result}")
except ZeroDivisionError:
    print("Error: You cannot divide by zero!")
except ValueError:
    print("Error: Please enter a valid integer.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
finally:
    print("This block always runs, no matter what.")
````

### 12.3 List Comprehensions
This is a concise way to create lists from existing iterables. It's often faster and more 'Pythonic' than using a standard `for` loop.

````python
numbers = [1, 2, 3, 4, 5, 6]
# Create a list of squares for even numbers only
even_squares = [x**2 for x in numbers if x % 2 == 0]
print(even_squares) # Output: [4, 16, 36]
````

### 12.4 Enumerate & Global
- **Enumerate**: Gives you both the index and the value of items in a list during a loop.
- **Global**: Allows you to modify a variable defined outside the function.

````python
fruits = ["Apple", "Banana", "Cherry"]
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")

x = 10
def change_global():
    global x
    x = 20

change_global()
print(f"Value of x: {x}")
````
